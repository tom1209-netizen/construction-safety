# Fine Tune YOLOv10

## Download the dataset

```python
%cd /content

# Download the dataset
!gdown 'https://drive.google.com/uc?id=1twdtZEfcw4ghSZIiPDypJurZnNXzMO7R' -O 'Safety_Helmet_Dataset.zip'

# Create the directory
!mkdir safety_helmet_dataset

# Unzip the dataset into the created directory
!unzip -q 'Safety_Helmet_Dataset.zip' -d '/content/safety_helmet_dataset'
```

## Train YOLO

```python
YAML_PATH = '../safety_helmet_dataset/data.yaml'
EPOCHS = 100
IMG_SIZE = 640
BATCH_SIZE = 128

model.train(data=YAML_PATH,
            epochs=EPOCHS,
            batch=BATCH_SIZE,
            imgsz=IMG_SIZE)
```

## Evaluate the model

```python
TRAINED_MODEL_PATH = 'runs/detect/train/weights/best.pt'
model = YOLOv10(TRAINED_MODEL_PATH)

model.val(data=YAML_PATH,
          imgsz=IMG_SIZE,
          split='test')
```
